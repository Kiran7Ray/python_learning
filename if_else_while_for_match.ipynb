python if statement

```python
#eg
a=23
b=300
if b>a:
    print(f"{b} is greater than {a}")
    print(f"{a} is smaller than {b}")

#using variable in condition
value=True
if value:
    print("the value is true")
```

```python
#elif keyword
a=23
b=20
if b>a:
   print(f"{b} is greater than {a}")
elif a>b:
   print(f"{a} is greater than {b}")

#multiple elif
num=10
if num==0:
    print("the number is zero")
elif num>0:
    print("the num is positive")
elif num<0:
   print("the num is negative")
          
```

```python
#else statement
a=200
b=200
if a>b:
     print(f"{a} is greater than {b}")
elif b>a:
     print(f"{a} is greater than {b}")
else:
     print(f"{a} equal to {b}")
```

```python
#short hand if
a=5
b=2
if a>b: print(f"{a} is greater than {b}")

#short hand if....else
item=1
print("FULL") if item==1 else print("Empty")

#assign a value with if...else
##variable = value_if_true if condition else value_if_false
a=10
b=20
bigger=a if a>b else b
print(f"bigger is {bigger}")

#multiple condition on one line
a=300
b=222
print("A") if a>b else print("=") if a==b else print("B")
```

```python
#nested if
age=18
licence=True

if age>=18:
    if licence:
        print("CAN DRIVE")
    if not licence:
        print("CAN NOT DRIVE")
else:
    print("YOU ARE UNDER AGE")

#pass vs comments
root=-20
if root<0:
    pass #added later
elif root>0:
    print("root possible")
else:
    print("root is zero")
```

#Python match statement
match expression:
  case x:
    code block
  case y:
    code block
  case z:
    code block

```python
day=5
match day:
    case 1:
        print("sunday")
    case 2:
        print("monday")
    case 3:
        print("tuesday")
    case 4:
        print("wednesday")
    case 5:
        print("thursday")
    case 6:
        print("friday")
    case 7:
        print("saturday")
    case _:                  #default value
        print("error input")  
```

```python
#combine value
day = 4
match day:
  case 1 | 2 | 3 | 4 | 5:
    print("Today is a weekday")
  case 6 | 7:
    print("I love weekends!")
```

```python
#python while loop
i=1
while i<6:
    print(i)
    i+=1

j=1
while j<6:
   print(j,end=" ")
   if j==4:
       break      #break statement we can stop the loop
   j+=1

k = 1
while k < 6:
    if k == 4:
      continue  #continue â†’ skips the rest of the loop body
    print(k, end=" ")
    k += 1

```

```python
#python for loop

fruits=['a','b','c','d','e']
for x in fruits:
   print(x)

# loop through string
for i in "school":
    print(i,end=" ")

for x in range(6):
    print(x)

for x in range(2, 6):
  print(x,end= " ")
```
