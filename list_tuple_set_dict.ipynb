Python list
-list stores multiple items in a single variable.
-list items are ordered,changeable and allow duplication

```python
my_list=[1,"apple",2.33,"school",1]
print(my_list)
print(my_list[2]) #ordered
print(len(my_list)) #list length

#list constructor
new_list=list(("apple","oranges","banana","coconut"))
print(new_list)
```

```python
#indexing-first item has index 0
my_list=[1,"apple",2.33,"school",1]
print(my_list[1])
#negative indexing
print(my_list[-2])

#range
print(my_list[1:4])
print(my_list[:4])
print(my_list[3:])

#check if item exist or not
new_list=list(("apple","oranges","banana","coconut"))

if 'apple' in new_list:
    print("apple lies in the list")
```

```python
#changing value
list1=[2,4,6,8,"ten"]
print(list1)
list1[4]=10
print(list1)

list1[2:4]=["four","six"]
print(list1)

list1[:2]=["even"] #changes index 0 & 1 value by one value
print(list1)
```

```python
#insert item
list2=["java","python","c","c++"]
list2.insert(3,"c#") #insert(index,value)
print(list2)

list2.append("javascript") #append(value)
print(list2)

#extend  list
li1=[1,2,3,4]
li2=["one","two","three",4]
print(li1)
li1.extend(li2) #extend(list)
print(li1)
print(li2)
```

```python
#remove items
items=["java","python","c","c++"]
items.remove("c") #remove(value)
print(items)
items.remove("python")
print(items)

my_list=[1,"apple",2.33,"school",10]
my_list.pop(2)  #pop(index)
print(my_list) 
my_list.pop() #remove last item
print(my_list)

del my_list[1] #del remove specified index
print(my_list)
del my_list #del completely remove the list

list_new=[1,10,100,"thousand"]
list_new.clear() #clear empty the list
print(list_new)
```

```python
#looping through the list
list1=["apple", "banana", "cherry"]
#Print all items in the list, one by one:
for x in list1:
    print(x)
#Print all items by referring to their index number:
for i in range(3):
    print(list1[i])
```

```python
#list comprehension
'''List comprehension offers a shorter syntax when you want to 
create a new list based on the values of an existing list'''
list_fruits=["mango","orange","apple","cherry","kiwi"]
newlist=[]
for x in list_fruits:
    if "a" in x:
        newlist.append(x)

print(newlist)
#With list comprehension you can do all that with only one line of code:
'''The Syntax
newlist = [expression for item in iterable if condition == True]
'''
list1=[x for x in list_fruits if "a" in x]
print(list1)
```

```python
#list sort // case sensitive return all capital letter being sort before lower case letter
thislist = ["orange", "mango", "kiwi", "pineapple", "banana"]
thislist.sort() #sort alphabetically
print(thislist)

thislist.sort(reverse=True)
print(thislist)

# case-insensitive sort function, use str.lower as a key function
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort(key = str.lower)
print(thislist)

#reverse order
newlist=[1,2,3,4,5,6]
newlist.reverse()
print(newlist)
```

```python
#copy a list
list1=["fire","wood","hot","smoke"]
list2=list1.copy()
print(list2)
#list2==list1 can be done but changes made on list1 will affect list2
list3=[]
list3=list1
list3[0]="water"
print(list1)

#use list() method
list4=["school","books","students","college"]
list5=list(list4)
print(list5)

#using slice operator
list6=list4[:]
print(list6)
```

```python
#joining two lists
list_1=["a","e","i","o","u"]
list_2=[1,2,3,4,5]
list_3=list_1+list_2
print(list_3)

#or

for x in list_2:
    list_1.append(x)

print(list_1)

#or use extend()
```

Tuples
store multiple items in a single variable.
ordered,allow duplicate  and unchangeable

```python
tuple_new=("apple","ball","cat","dog",1,"cat")
print(tuple_new)
#ordered / indexing
print(tuple_new[3])
print(tuple_new[:3])
print(tuple_new[2:5]) #range
print(tuple_new[-2])
#tuple constructor
tuple1=tuple(("apple","ball","cherry"))
print(tuple1)
print(type(tuple1))
```

```python
#unchangeable
x=("school","college","student")
print(x[1])
# x[1]="teacher" x.append("teacher") #not possible

#convert to list then do it
y=list(x)
y[2]="teacher"
y.append("library")
y.remove("college")
x=tuple(y)
print(x)

```

```python
#unpacking tuples
#we create a tuple, we normally assign values to it. This is called "packing" a tuple
fruits=("apple","banana","cherry")
#unpacking a tuple
(x,y,z)=fruits
print(x)
print(y)
print(z)
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")
#using *
numbers=(1,3,5,7,10,12,14)
(x,y,*z)=numbers
print(x)
print(y)
print(z)
print(type(z))
```

```python
#loop through tuple
tuple1=("box","socks","shoes")
for x in tuple1:
    print(x)
```

```python
#joining tuple
tp1=("a","b","c")
tp2=(1,2,3)
tp3=tp1+tp2
print(tp3)

tp4=tp1*2
print(tp4)
```

python sets
used to store multiple items in a single variable.
set is a collection which is unordered, unchangeable*, and unindexed.
Duplicates Not Allowed

```python
my_set={"copy","book","box","pen",1,True}
print(my_set)
#True and 1 ,False and 0 is considered the same value
print(len(my_set))
```

```python
#set constructor
new_set=set((1,"apple","good",False))
print(type(new_set))
```

```python
for x in  new_set:
    print(x)

print("apple" in new_set)
```

```python
#adding items
set1={"apple","banana","mango"}
set1.add("orange")
print(set1)

set2={1,2,3}
set2.update(set1)
print(set2)

#removing
set2.remove(1) #remove raise error if value say 1 to remove is not present in set
print(set2)
set2.discard("apple") #discard does not raise error
print(set2)

set2.pop()
print(set2)

set2.clear() #empty the set
print(set2)

del set2 #delete the set completely
```

```python
for x in set1:
    print(x)
```

```python
#join sets
set1={"a","i","o","w"}
set2={"e","u",1,"r","a"}
#union() method and | operatur does same
set3=set1.union(set2)
print(set3)

set4=set1|set2
print(set4)
#multiple set
set11={"x","y"}
set3=set1.union(set2,set11) #or set3=set1|set2|set11
print(set3)

#intersection ,&
setnew=set1.intersection(set2)
print(setnew)

set_new=set1&set2
print(set_new)
# intersection_update() method will also keep ONLY the duplicates, but it will change the original set instead of returning a new set.
set1.intersection_update(set2)
print(set1)

#difference , -
set12=set2.difference(set1) #set2-set1
print(set12)
print(set2-set1)
#difference_update() method to keep only the items from the first set that are not present in the other set:
set2.difference_update(set1)
print(set2)
#symmetric_difference() method will keep only the elements that are NOT present in both sets.
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set3 = set1.symmetric_difference(set2)
print(set3)

set3=set1^set2
print(set3)
#Use the symmetric_difference_update() method to keep the items that are not present in both sets:
set1 = {"apple", "banana", "cherry"}
set2 = {"google", "microsoft", "apple"}

set1.symmetric_difference_update(set2)

print(set1)
```

```python
#frozenset- set that cannnot be innutable
x=frozenset({"box","ball","bed"})
print(x)
```

dictionary
-key:value pairs
-ordered*, changeable and do not allow duplicates.

```python
my_dict={
"faculty":"bct",
"roll_no":18,
"year":2080
}
print(my_dict)
#The values in dictionary items can be of any data type:
my_dict={
    "faculty":"bct",
"roll_no":18,
"year":2080,
    "pass":True,
    "subjects":["math","nepali","english"]
}
print(my_dict)
print(type(my_dict))

#dict() constructor
new_dict=dict(name='kiran',age=18,country="nepal")
print(new_dict)
```

```python
#accessing value
x=new_dict['name']
print(x)

x=new_dict.get('country')
print(x)

x=new_dict.keys()
print(x)

x=new_dict.values()
print(x)
```

```python
#changing values
new_dict={
"country":"nepal",
"district":"ilam",
"dob":2004
}
print(new_dict)
new_dict['dob']=2000
print(new_dict)

new_dict.update({'dob':2004})
print(new_dict)

#adding value
new_dict['color']="blue"
print(new_dict)

new_dict.update({"fruit":"mango"})
print(new_dict)
```

```python
#remove items
new_dict.pop("fruit")
print(new_dict)

# new_dict.popitem()
# print(new_dict)

del new_dict["dob"]
print(new_dict)

new_dict.clear()
print(new_dict)
```

```python
#loop through dictionary
dictA={
    "animal":"cat",
    "number":20,
    "college":"nce",
    "correct":True,
    "fruit":"orange"
}
print(dictA)

for x in dictA:
    print(x)

for x in dictA.values():
    print(x)

for x in dictA.keys():
    print(x)

for x,y in dictA.items():
    print(x,y)
```

```python
#copy dictionary
dict1={
    "a":"apple",
    "b":"ball",
    "c":"cat"
}
dict2=dict1.copy()
print(dict2)

dict3=dict(dict1)
print(dict3)
```
