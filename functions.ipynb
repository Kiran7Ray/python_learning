```python
#Positional-only arguments are function parameters that must be given by position, not by name.
# Example Function Definition
def greet(name,/, greeting="Hello"): # 'name' is positional-only, 'greeting' is standard
    print(f"{greeting}, {name}!")

# Valid Calls (Positional-Only 'name')
greet("Alice")         # Works: 'name' gets "Alice" by position
greet("Bob", "Hi")     # Works: 'name' gets "Bob", 'greeting' gets "Hi"

# Invalid Call (Keyword for 'name')
# greet(name="Charlie") # TypeError: greet() got some positional-only arguments passed as keyword arguments

#Combining Positional-Only and Keyword-Only
def my_function(a, b, /, *, c, d):
  return a + b + c + d

result = my_function(5, 10, c = 15, d = 20)
print(result)
```

*args and **kwargs
By default, a function must be called with the correct number of arguments.

However, sometimes you may not know how many arguments that will be passed into your function.

*args and **kwargs allow functions to accept a unknown number of arguments.

```python
#Arbitrary Arguments - *args
#*args parameter allows a function to accept any number of positional arguments.
def my_function(*kids):
  print("The youngest child is " + kids[2])

my_function("Emil", "Tobias", "Linus")

def my_function1(*args):
  print("Type:", type(args))
  print("First argument:", args[0])
  print("Second argument:", args[1])
  print("All arguments:", args)

my_function1("Emil", "Tobias", "Linus")
```

```python
# **kwargs parameter allows a function to accept any number of keyword arguments.
# Inside the function, kwargs becomes a dictionary containing all the keyword arguments:

def my_function(**myvar):
  print("Type:", type(myvar))
  print("Name:", myvar["name"])
  print("Age:", myvar["age"])
  print("All data:", myvar)

my_function(name = "Tobias", age = 30, city = "Bergen")

def my_function2(username, **details):
  print("Username:", username)
  print("Additional details:")
  for key, value in details.items():
    print(" ", key + ":", value)

my_function2("emil123", age = 25, city = "Oslo", hobby = "coding")
```

```python
#Combining *args and **kwargs
def my_function(title, *args, **kwargs):
  print("Title:", title)
  print("Positional arguments:", args)
  print("Keyword arguments:", kwargs)

my_function("User Info", "Emil", "Tobias", age = 25, city = "Oslo")
```

```python
# python decorators
 # basic decorators
# Define the decorator first, then apply it with @decorator_name above the function.
#decorators is a function that take another function as argument and returns a function.
def changecase(func):
    def inner():
      return func().upper()
    return inner
@changecase
def function11():
    return "hello kiran"

print(function11())
###############################
def decorator(func):
    def innerfun():
        print("transaction proceed")
        func()
        print("transaction complete")
    return innerfun
    
@decorator
def hello():
    print("transaction continuing...")

hello()
```

```python
#Multiple Decorator Calls
def change(funct):
    def decorated():
        text=funct()
        print(text.upper())
    return decorated

@change
def low():
    return "HELLO kiran"

@change
def high():
    return "hello KIRAN"


low()
high()
```

```python
#Arguments in the Decorated Function
def decorate(funct1):
    def funct2(a,b):
        if a>b:
            a,b=b,a
            return funct1(a,b)
    return funct2
    
@decorate
def divide(a,b):
  return b/a  #we want b>a

print(divide(8,4))
```

```python
# python lambda
# lambda arguments : expression

x=lambda a:a+10
print(x(5))

y=lambda a,b,c:a+b+c
print(y(3.4,34,5))

#The power of lambda is better shown 
#when you use them as an anonymous function inside another function.
def my_function(n):
    return lambda a:a*n

value=my_function(20)   #value=lambda a:a*20
print(value(10))
```

```python
#Lambda with Built-in Functions
#map() function applies a function to every item in an iterable:
num=[1,2,3,4,5]
double=list(map(lambda x:x*2,num))
print(double)
#filter() function creates a list of items for which a function returns True:
num2=[1,2,3,4,5,6,7,8]
odd_num=list(filter(lambda x:x%2!=0,num2))
print(odd_num)
#sorted() function can use a lambda as a key for custom sorting:
student=[("kiran",200),('rishav',10),('ram',68)]
sort_std=sorted(student,key=lambda x:x[1])
print(sort_std)
```

```python
#Python Recursion
def countdown(n):
  if n <= 0:
    print("Done!")
  else:
    print(n)
    countdown(n - 1)

countdown(5)
```

```python
# Generators
# Generators are functions that can pause and resume their execution.
def my_generator():
  yield 1
  yield 2
  yield 3

for value in my_generator():
  print(value)

# Generators allow you to iterate over data without storing the entire dataset in memory.
# Instead of using return, generators use the yield keyword.
# When yield is encountered, the function's state is saved, and the value is returned. 
#The next time the generator is called, it continues from where it left of

# Using next() with Generators
# You can manually iterate through a generator using the next() function:
def simple_gen():
  yield "Emil"
  yield "Tobias"
  yield "Linus"

gen = simple_gen()
print(next(gen))
print(next(gen))
print(next(gen))
```
