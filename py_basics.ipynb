python string

```python
a="python string"
print(a)
#string are array
print(a[2])
print(a[5])
#looping through sting
for x in 'python':
    print(x)
#check_string
text="the best thing is to learn python"
print("python" in text)

#use if statement
if "best" in text:
    print("yes python is best")

#use not in ststement
print("java" not in text)
if "java" not in text:
    print("we are not learning java")

#slicing
x="hello python"
print(x[2:5])

print(x[:3]) #slice from start
print(x[2:]) #slice from end

print(x[-5:-2]) #negative slicing

#upper case + lower case
a="waterisfluid"
print(a.upper())
print(a.lower())

#remove white space
b="hello in python"
print(a.strip()) #strip() method removes any whitespace

#replace string
a="replace string"
print(a.replace('r','h')) #replace() method replaces a string with another string

#split string
#split() method returns a list where the text between the specified separator becomes the list items.
a="hello ,there"
print(a.split(","))

#string concatenation
a="hello"
b="world"
c=a+b
print(c)
print(a+" "+b)

#string_format
#f_string
age=20
txt=f"my name is kiran , i am {age}"
print(txt)
```

python operator

```python
#arithmetic operator
print(2+3) #addition
print(5-2) #substraction
print(2*2) #multiplication
print(2**3) #power
print(4/3) #division
print(3%2) #modulus
print(15//2) #floor division

#assidnment operator
x=5
x+=3
x-=2
x*=2
x/=3
x**=3
print(x)
#Walrus Operator , :=
print(y:=4) #y=4 and print 4

#comparator operators
print(5==5) #equal
print(2!=3) #not equal
print(2<3) #less then
print(3>2) #greater then
print(3>=3) #greater then or equal to
print(3<=5) #less then or equal to

#logical operator
x=5
print(x==5 and x<20) #and
print(x<10 or x<20) #or
print(not(x==5)) #not

#identity operator
x=['good','great','bad']
z=['good','great','bad']
print(x is z) #	Returns True if both variables are the same object
print(x is not z)
print(x==z)
'''is - Checks if both variables point to the same object in memory
== - Checks if the values of both variables are equal
'''

#membership operator
fruit=['mango','apple','orange']
print('banana' in fruit) #Returns True if a sequence with the specified value is present in the object
print('mango' in fruit)
print('coconut' not in fruit) #Returns True if a sequence with the specified value is not present in the object

#bitwise operators-compare binary operators
print(5&3) #5=>101 (and) 3=>011 =>001(1)
print(6|3) #or
print(6^2) #xor
print(~7) #not
print(3<<2)#	Zero fill left shift (shift binary 3,two times left)
print(3<<2) 
```
